## Retention Rate ##

```python
# Obtenci√≥n de la primera fecha de visita
visits['date_st'] = pd.to_datetime(visits['date_st'])
first_act_date = visits.groupby('Uid')['date_st'].min()
first_act_date.name = 'first_act_date'
first_act_date.head()
```

```python
visits = visits.join(first_act_date, on = 'Uid')
visits.head()
```

```python
# Obtenci√≥n del primer mes de visita
visits['first_act_month']=visits['first_act_date'].dt.to_period('M')
visits['act_month']=visits['date_st'].dt.to_period('M')
visits.head()
```

```python
# Obtenci√≥n de cohortes por mes
#visits['cohort_lt'] = (visits['date_st'] - visits['first_act_date'])
visits['cohort_lt'] = (visits['act_month'].dt.year - visits['first_act_month'].dt.year)*12 + (visits['act_month'].dt.month - visits['first_act_month'].dt.month) 
visits.head()
```

```python
cohorts = (visits.groupby(['first_act_month', 'cohort_lt'])
    .agg({'Uid': 'nunique'})
    .reset_index())
cohorts
```

```python
# Usuarios unicos por primer mes y mes de actividad
#cohort_users = visits.groupby(['first_order_month', 'act_month']).agg({'Uid': 'nunique'}).reset_index()
#cohort_users.columns = ['first_order_month', 'act_month', 'n_users']
#cohort_users
```

```python
# N√∫mero inicial de usuarios por cohorte
initial_users_count = cohorts[cohorts['cohort_lt']==0][['first_act_month', 'Uid']]
initial_users_count = initial_users_count.rename(columns={'Uid': 'cohort_users'})
initial_users_count
```

```python
cohorts = cohorts.merge(initial_users_count,on='first_act_month')
cohorts
```

```python
# Tasa de retenci√≥n
cohorts['retention'] = cohorts['Uid']/cohorts['cohort_users']
cohorts.head()
```

```python
retention_pivot = cohorts.pivot_table(
    index='first_act_month',
    columns='cohort_lt',
    values='retention',
    aggfunc='sum',
)
retention_pivot.head()
```

```python
sns.set(style='white')
plt.figure(figsize=(13, 9))
plt.title('Cohortes: retenci√≥n de usuarios/as')
sns.heatmap(
    retention_pivot, annot=True, fmt='.1%', linewidths=1, linecolor='gray'
)
```

Aunque parecia que las sesiones diarias iban en incremento, los usuarios no estan regresenado al sitio despues de una primera visita.

## AGRUPACI√ìN DE COHORTES ##

```python
# Agrupaci√≥n por meses
cohort_orders = orders.groupby(['first_order_month', 'order_month']).agg({'Revenue': 'sum', 'Uid': 'nunique'}).reset_index()
cohort_orders.columns = ['first_order_month', 'order_month', 'total_revenue', 'n_users']
```

```python
# Ingreso pos usuario por cohorte
cohort_orders['cohort_lifetime'] = (cohort_orders['order_month'] - cohort_orders['first_order_month']).apply(lambda x: x.n)
cohort_orders['revenue_per_user'] = cohort_orders['total_revenue'] / cohort_orders['n_users']
cohort_orders.head()
```

```python
pivot = cohort_orders.pivot_table(
    index='first_order_month',
    columns='cohort_lifetime',
    values='revenue_per_user',
    aggfunc='mean',
    fill_value=0
)
```

```python
plt.figure(figsize=(12, 6))
sns.heatmap(pivot, annot=True, fmt=".2f", cmap="YlGnBu", linewidths=1, linecolor='gray')
plt.title("Ingresos promedio por usuario seg√∫n cohorte y mes de vida")
plt.xlabel("Meses desde primera compra")
plt.ylabel("Mes de primera compra (cohorte)")
plt.tight_layout()
plt.show()
```

```python
# Usuarios √∫nicos por fuente
user_by_source = visits.groupby('Source Id')['Uid'].nunique()
user_by_source
```

```python
# Generaci√≥n de tabla con fuente, usuarios y meses
user_by_source = costs.merge(user_by_source, left_on ='source_id', right_on ='Source Id', how = 'right')
user_by_source
```

```python
# Costoss totales por fuente
#costs_by_source = costs.groupby('source_id')['costs'].sum()
#costs_by_source
```

```python
# Agrupaci√≥n de costos por canal por mes
costs_by_month_channel = costs.groupby(['monthly', 'source_id'])['costs'].sum().reset_index()
costs_by_month_channel.columns = ['cohort', 'source_id', 'total_cost']
costs_by_month_channel
```

```python
orders.info()
```

```python
# Agrupar pedidos por usuario 
first_order = orders[['Uid', 'first_order_month']]
first_order
```

```python
first_order_visits = first_order.merge(visits, on='Uid', how='left')
```

```python
# Agrupaci√≥n por de usuarios por mes y por canal
users_by_cohort_channel = first_order_visits.groupby(['first_order_month', 'Source Id'])['Uid'].nunique().reset_index()
users_by_cohort_channel.columns = ['cohort', 'source_id', 'n_users']
users_by_cohort_channel
```

```python
cac_df = pd.merge(costs_by_month_channel, users_by_cohort_channel, on=['cohort', 'source_id'], how='inner')
cac_df['CAC'] = cac_df['total_cost'] / cac_df['n_users']
cac_df
```
